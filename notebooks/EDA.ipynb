```python
df = pl.read_parquet('../data/clean_training_data.parquet')
df
```

## Univariate Analysis
This section goes through the avilable columns and plots them to see the distributions, outliers, etc. 

```python
fig = px.histogram(df, x='ip_type', width=600, height=400,)
# remove beckground grid
fig.update_layout(
    title='Distribution of IP Types',
    xaxis_title='IP Type',
    yaxis_title='Count',
    template='simple_white'
)
fig.show()
```

```python
fig = px.histogram(df, x='source', width=600, height=400,)
# remove beckground grid
fig.update_layout(
    title='Distribution of source',
    xaxis_title='source',
    yaxis_title='Count',
    template='simple_white'
)
fig.show()
```

```python
fig = px.histogram(df, x='classification', width=600, height=400,)
# remove beckground grid
fig.update_layout(
    title='Distribution of classification',
    xaxis_title='classification',
    yaxis_title='Count',
    template='simple_white'
)
fig.show()
```
