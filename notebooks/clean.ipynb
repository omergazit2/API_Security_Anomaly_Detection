## Data pre-processing pipline

Based on the anaysis above, we take the following data cleaning steps:
1. remove rows with null values. 
2. Drop the id ids column.
3. The file will be saved as parquet file with no numerical id index.

```python
def clean_data(df: pl.DataFrame, save_path: str = '../data/clean_training_data.parquet') -> None:
    """Clean the data and save it as a parquet file.
    Args:
        df (pl.DataFrame): the dataframe to clean
        save_path (str): the path to save the cleaned dataframe
    
    Returns:
        None
    """
    df = df.filter(pl.any_horizontal(pl.all().is_not_null()))
    df.write_parquet(save_path)
```

```python
clean_data(df)
```
